```python
import pandas as pd
import matplotlib.pyplot as plt                                                                                                                                                                                                          
import pylab
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import cross_val_score
```

```python
linear1 = pd.read_csv("linear_jumal.tsv", sep='\t')
X = linear1[['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10']]
y = linear1[['SCORE']]
linearReg1 = LinearRegression().fit(X, y)
scores = cross_val_score(linearReg1, X, y, cv=5, scoring='r2')
scores.mean(), scores.std()
```

```python
logistic1 = pd.read_csv("logistic_salmyeonseo.tsv", sep='\t')
X = logistic1[['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10']]
y = logistic1[['LEVEL']]
clf = LogisticRegressionCV(cv=5, random_state=0).fit(X, y)
scores = cross_val_score(clf, X, y, cv=5)
scores.mean(), scores.std()
```

```python
from sklearn import svm
nli = pd.read_csv("nli.tsv", sep='\t')
nli = nli.dropna()
X = nli[['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10']]
y = nli[['NATIVE']]
clf = svm.SVC(kernel='linear', C=1, random_state=42)
scores = cross_val_score(clf, X, y, cv=5)
scores.mean(), scores.std()
```
